```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import os
```

```python
PATH = "Raport/"
results = []
results_names = []
for el in os.listdir(PATH):
    if el[-21:-4] == 'evaluation_values':
        results_names.append(el)
        results.append(np.load(PATH + el)['arr_0'])
```

```python
r = results[0]
plt.plot(np.mean(r, axis=0))
plt.title(results_names[0])
print("min  :", np.min(r))
print("mean :",np.mean(r))
print("std  :",np.std(r),"\n")
df_describe = pd.DataFrame(r)
df_describe.describe()
```

```python
plt.hist(np.mean(r, axis=0), bins=10, color='c', edgecolor='k', alpha=0.5)
```

```python
stats.ttest_ind(r[0], r[1],equal_var = False)
```

```python
stats.wilcoxon(r[0], r[1])
```
